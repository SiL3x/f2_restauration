```python
data_f =  [50, 100, 150, 200, 300, 400, 500, 750, 1000, 1250, 1500, 2000, 2500, 3000, 4000, 5000, 7500, 10000, 12000]  # f[Hz]
data_ht = [2.0,2.6, 2.6, 2.2, 2.4, 2.8, 3.0, 3.6,  3.8,  3.8, 3.76,  3.6, 3.28, 3.12, 3.04,  3.6,  5.4,  11.4,  10.1]  # Uss[V]
data_mt = [5.0,7.0, 6.4, 5.4, 5.4, 6.2, 6.2, 5.4,  4.0,  2.9,  2.2, 1.44, 0.96, 0.72, 0.48, 0.36, 0.12,  0.11,  0.17]  # Uss[V]
```
