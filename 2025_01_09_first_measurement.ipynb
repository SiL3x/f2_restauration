```python
import matplotlib.pyplot as plt
```

```python
# Proof of Principle Messung

data_f =  [50, 100, 150, 200, 300, 400, 500, 750, 1000, 1250, 1500, 2000, 2500, 3000, 4000, 5000, 7500, 10000, 12000]  # f[Hz]
data_ht = [2.0,2.6, 2.6, 2.2, 2.4, 2.8, 3.0, 3.6,  3.8,  3.8, 3.76,  3.6, 3.28, 3.12, 3.04,  3.6,  5.4,  11.4,  10.1]  # Uss[V]
data_mt = [5.0,7.0, 6.4, 5.4, 5.4, 6.2, 6.2, 5.4,  4.0,  2.9,  2.2, 1.44, 0.96, 0.72, 0.48, 0.36, 0.12,  0.11,  0.17]  # Uss[V]
```

```python
plt.yscale("log")
plt.xscale("log")
plt.xlabel("f [Hz]")
plt.ylabel("Uss [V]")
plt.plot(data_f, data_ht, label="HT")
plt.plot(data_f, data_mt, label="MT")
plt.legend(loc="lower left")
plt.savefig("2025_01_09_pof_measurement.png")
```

```python
data = pd.read_csv("2025_01_09_amp.csv", sep="\t")
print(data.columns)
data_f = list(data["f"])
data_amp_1_cut_ht = list(data["ht cut"])
data_amp_1_cut_mt = list(data["mt cut"])
data_amp_1_flat_ht = list(data["ht flat"])
data_amp_1_flat_mt = list(data["mt flat"])
data_amp_1_boost_ht = list(data["ht boost"])
data_amp_1_boost_mt = list(data["mt boost"])

data_amp_2_cut_ht = list(data["ht cut.1"])
data_amp_2_cut_mt = list(data["mt cut.1"])
data_amp_2_flat_ht = list(data["ht flat.1"])
data_amp_2_flat_mt = list(data["mt flat.1"])
data_amp_2_boost_ht = list(data["ht boost.1"])
data_amp_2_boost_mt = list(data["mt boost.1"])
```

```python
plt.yscale("log")
plt.xscale("log")
plt.xlabel("f [Hz]")
plt.ylabel("Uss [V]")
plt.plot(data_f, data_amp_1_cut_ht, label="Cut HT")
plt.plot(data_f, data_amp_1_cut_mt, label="Cut MT")
plt.plot(data_f, data_amp_1_flat_ht, label="Flat HT")
plt.plot(data_f, data_amp_1_flat_mt, label="Flat MT")
plt.plot(data_f, data_amp_1_boost_ht, label="Boost HT")
plt.plot(data_f, data_amp_1_boost_mt, label="Boost MT")
plt.legend(loc="lower left")
plt.savefig("2025_01_10_measurement_amp_1.png")
```
